```python
# importing the necessary liberies

from flask import Flask, render_template, request, jsonify
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as urReq
import company_urls as cu
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Final Code for finding the variables of company
    
# https://www.moneycontrol.com/india/stockpricequote/
    
if __name__ == "__main__":
    a_urls = cu.company_urls()
    
    
company_url = a_urls[0:len(a_urls)]
Company_Names = []
Year = []
Profit_before_tax = []
Finance_Costs = []
Depreciation = []
EBITDA = []
Revenue_From_Operations = []
Purchase_Of_Stock_In_Trade = []
Cost_Of_Materials_Consumed = []
Changes_In_Inventories = []
Gross_Profit = []
Cost_of_Goods_Sold = []
Employee_Benefit_Expenses = []
Depreciation_And_Amortisation_Expenses = []
Other_Expenses = []
EBIT = []



for l in company_url:
    
    try:
        company_page2 = requests.get(l)
        company_page2.encoding = "utf-8"
        beautified_company_page2 = bs(company_page2.text,"html.parser")
        detailed_financial2 = beautified_company_page2.find_all("div",{"class":"quick_links clearfix"})
        detailed_financial3 = detailed_financial2[0].find_all("li")

        profit_loss_statement = detailed_financial3[1].a["href"]
        profit_loss_statement_url = requests.get(profit_loss_statement)
        profit_loss_statement_url.encoding = "utf-8"
        beaurified_profit_loss_statement_url = bs(profit_loss_statement_url.text,"html.parser")

        company_name2 = beaurified_profit_loss_statement_url.find_all("div",{"class":"moneyprice_bx"})
        company2 = company_name2[0].h1.text


        profit_loss_statement_text2 = beaurified_profit_loss_statement_url.find_all("div", {"class":"table-responsive financial-table"})
        temp2 = profit_loss_statement_text2[0].find_all("tr")
        temp3 = temp2[21].find_all("td")
        temp4 = temp2[0].find_all("td")
        for n in range(1,(len(temp4[0:-1]))):
            PROFIT_BEFORE_TAX = temp3[n].text
            year2 = temp4[n].text
            
        temp3 = temp2[15].find_all("td")
        for n in range(1,(len(temp4[0:-1]))):
            FINANCE_COSTS = temp3[n].text           
            
        temp3 = temp2[16].find_all("td")
        for n in range(1,(len(temp4[0:-1]))):
            DEPRECIATION = temp3[n].text
        
        ebitda = round(float((float(PROFIT_BEFORE_TAX) + float(FINANCE_COSTS) + float(DEPRECIATION))),2)
        
        temp3 = temp2[6].find_all("td")
        for n in range(1,(len(temp4[0:-1]))):
            REVENUE_FROM_OPERATIONS = temp3[n].text
        
        temp3 = temp2[11].find_all("td")
        for n in range(1,(len(temp4[0:-1]))):
            PURCHASE_OF_STOCK_IN_TRADE = temp3[n].text
            
        temp3 = temp2[10].find_all("td")
        for n in range(1,(len(temp4[0:-1]))):
            COST_OF_MATERIALS_CONSUMED = temp3[n].text
        
        temp3 = temp2[13].find_all("td")
        for n in range(1,(len(temp4[0:-1]))):
            CHANGES_IN_INVENTORIES = temp3[n].text
        
        
        GROSS_PROFIT = round(float((float(REVENUE_FROM_OPERATIONS) - float(COST_OF_MATERIALS_CONSUMED) - 
                                    float(PURCHASE_OF_STOCK_IN_TRADE) - float(CHANGES_IN_INVENTORIES))),2)
        
        COST_OF_GOODS_SOLD = round(float((float(COST_OF_MATERIALS_CONSUMED) + float(PURCHASE_OF_STOCK_IN_TRADE) + 
                                          float(CHANGES_IN_INVENTORIES))),2)
        
        temp3 = temp2[14].find_all("td")
        for n in range(1,(len(temp4[0:-1]))):
            EMPLOYEE_BENEFIT_EXPENSES = temp3[n].text
            
            
        temp3 = temp2[16].find_all("td")
        for n in range(1,(len(temp4[0:-1]))):
            DEPRECIATION_AND_AMORTISATION_EXPENSES = temp3[n].text
            
        temp3 = temp2[17].find_all("td")
        for n in range(1,(len(temp4[0:-1]))):
            OTHER_EXPENSES = temp3[n].text
        
        
        ebit = round(float((float(PROFIT_BEFORE_TAX) + float(FINANCE_COSTS))),2)
        
        
        
        
        Company_Names.append(company2)
        Year.append(year2)
        Profit_before_tax.append(PROFIT_BEFORE_TAX)
        Finance_Costs.append(FINANCE_COSTS)
        Depreciation.append(DEPRECIATION)
        EBITDA.append(ebitda)
        Revenue_From_Operations.append(REVENUE_FROM_OPERATIONS)
        Purchase_Of_Stock_In_Trade.append(PURCHASE_OF_STOCK_IN_TRADE)
        Cost_Of_Materials_Consumed.append(COST_OF_MATERIALS_CONSUMED)
        Changes_In_Inventories.append(CHANGES_IN_INVENTORIES)
        Gross_Profit.append(GROSS_PROFIT)
        Cost_of_Goods_Sold.append(COST_OF_GOODS_SOLD)
        Employee_Benefit_Expenses.append(EMPLOYEE_BENEFIT_EXPENSES)
        Depreciation_And_Amortisation_Expenses.append(DEPRECIATION_AND_AMORTISATION_EXPENSES)
        Other_Expenses.append(OTHER_EXPENSES)
        EBIT.append(ebit)
        
        
    except Exception as e:
        print(e) 
            
    print(Company_Names)
    print(Year)
    print(Profit_before_tax)
    print(Finance_Costs)
    print(Depreciation)
    print(EBITDA)
    print(Revenue_From_Operations)
    print(Purchase_Of_Stock_In_Trade)
    print(Cost_Of_Materials_Consumed)
    print(Changes_In_Inventories)
    print(Gross_Profit)
    print(Cost_of_Goods_Sold)
    print(Employee_Benefit_Expenses)
    print(Depreciation_And_Amortisation_Expenses)
    print(Other_Expenses)
    print(EBIT)
    
```
