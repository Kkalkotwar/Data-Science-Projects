# 6.c Categorical to numerical Conversion

In this process all the categorical data is been converted into numerical form

# 6.c.i Using the Dummy Variable Trap technique

```python
# As in the above data we can see that there are so much of categorical data.
# By using the dummy variable trap techniqe we can convert the categorical data into numerical data.

df= pd.get_dummies(data=df,columns=['gender','PhoneService','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn'],drop_first= True )
df
```

# 6.c.ii Using Labeling Encoding for Numerical conversion

```python
# Now using the Label Encoder technique for  changing the categorical data to numerical 

from sklearn import preprocessing # Importing the preprocessing 

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()


df['MultipleLines']= label_encoder.fit_transform(df['MultipleLines'])
df['Contract']= label_encoder.fit_transform(df['Contract'])
df['PaymentMethod']= label_encoder.fit_transform(df['PaymentMethod'])
```

# 6.d Outlier Treatment

# 6.d.i Outlier Detection

```python
df.columns
```

```python
# Detection of outlier in the dataset

plt.figure(figsize=(20,10))
sns.boxplot(data = df, x = 'MonthlyCharges')
```

Now from the above graph it is not very much clear about the outlier is basically present or not. 
So we will try to deteck the outlire in the dataset using paercentile method

There are only three columns where we can see that the lagre value of data is present.
1. Tenure
2. MonthlyCharges
3. TotalCharges

So performing the percentile technique on these columns for checking that outlier is present or not


```python
max_threshold = df['MonthlyCharges'].quantile(0.99)   # Finding the maximum threshold value using percentile method
max_threshold
```

```python
upper_outlier=df[df['MonthlyCharges']>max_threshold]
upper_outlier
```

```python
upper_outlier.count()
```

```python
min_threshold = df['MonthlyCharges'].quantile(0.05)   # Finding the minimum threshold value using percentile method
min_threshold
```

```python
lower_outlier=df[df['MonthlyCharges']<min_threshold]
lower_outlier
```

```python
lower_outlier.count()
```

Now from the above outlier analysis based of monthly expense of a person, we can see that the upper_outlier (considered as 99%) is 53 in count. Where as in the lower_outlier (consideredderedd as 5%) is 276 in count.

----------------------------------------------------------------------------------------------------------------------------------------------------------------------

# 6.d.ii Outlier Removal

```python
# Outlier Removal 

Index_U = upper_outlier.index    # Getting the index values of the upper outliers
Index_U 
```

```python
Index_L = lower_outlier.index    # # Getting the index values of the lower outliers
Index_L
```

```python
# Droping the rows of the upper outliers
df.drop(Index_U, inplace=True)
```

```python
# Droping the rows of the lower outliers
df.drop(Index_L, inplace=True)
```
