```python
# importing the necessary liberies

from flask import Flask, render_template, request, jsonify
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as urReq
import company_urls as cu
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Final Code for finding the variables of company
    
# "https://www.moneycontrol.com/india/stockpricequote/"
    
if __name__ == "__main__":
    a_urls = cu.company_urls()
    
    
company_url = a_urls[0:len(a_urls)]
Company_Names = []
Total_Current_Liabilities = []
Total_Current_Assets = []
Year = []
Inventories = []
Cash_and_Cash_Equivalents = []
Current_Investments = []
Short_Term_Borrowings = []
Total_Assets = []
Total_Liabilities = []
Total_Non_Current_Assets = []
Total_Non_Current_Liabilities = []
Intangible_Assets = []
Fixed_Assets = []
Capital_Employed = []
Working_Capital = []
Trade_Receivables = []
Total_Equity = []
Trade_Payables = []



for l in company_url:
    
    try:
        company_page2 = requests.get(l)
        company_page2.encoding = "utf-8"
        beautified_company_page2 = bs(company_page2.text,"html.parser")
        detailed_financial2 = beautified_company_page2.find_all("div",{"class":"quick_links clearfix"})
        detailed_financial3 = detailed_financial2[0].find_all("li")

        balance_sheet2 = detailed_financial3[0].a["href"]
        balance_sheet_url2 = requests.get(balance_sheet2)
        balance_sheet_url2.encoding = "utf-8"
        beaurified_balance_sheet_url2 = bs(balance_sheet_url2.text,"html.parser")

        company_name2 = beaurified_balance_sheet_url2.find_all("div",{"class":"moneyprice_bx"})
        company2 = company_name2[0].h1.text


        balance_sheet_text2 = beaurified_balance_sheet_url2.find_all("div", {"class":"table-responsive financial-table"})
        temp2 = balance_sheet_text2[0].find_all("tr")
        temp3 = temp2[20].find_all("td")
        temp4 = temp2[0].find_all("td")
        for n in range(1,(len(temp4[0:-4]))):
            TOTAL_CURRENT_LIABILITIES=temp3[n].text      # Calculating the total current liabilities
            year2 = temp4[n].text
            
        temp3 = temp2[41].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            TOTAL_CURRENT_ASSETS=temp3[n].text           # Calculating the total current asset
        
        temp3 = temp2[36].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            INVENTORIES=temp3[n].text                    # Calculating the Inventories
        
        temp3 = temp2[38].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            CASH_AND_CASH_EQUIVALENTS=temp3[n].text
            
        temp3 = temp2[35].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            CURRENT_INVESTMENTS=temp3[n].text
        
        temp3 = temp2[16].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            SHORT_TERM_BORROWINGS=temp3[n].text
        
        temp3 = temp2[42].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            TOTAL_ASSETS=temp3[n].text
        
        
        temp3 = temp2[21].find_all("td")         # This is basically total debt
        for n in range(1,(len(temp4[0:-5]))):
            TOTAL_LIABILITIES=temp3[n].text
        
        
        temp3 = temp2[33].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            TOTAL_NON_CURRENT_ASSETS=temp3[n].text
        
        temp3 = temp2[14].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            TOTAL_NON_CURRENT_LIABILITIES=temp3[n].text
        
        temp3 = temp2[25].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            INTANGABLE_ASSETS=temp3[n].text
        
        temp3 = temp2[28].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            FIXED_ASSETS=temp3[n].text
                    
        CAPITAL_EMPLOYED = round(float((float(TOTAL_ASSETS) - float(TOTAL_CURRENT_LIABILITIES))),2)
        WORKING_CAPITAL = round(float((float(TOTAL_CURRENT_ASSETS) - float(TOTAL_CURRENT_LIABILITIES))),2)
        
        temp3 = temp2[37].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            TRADE_RECEIVABLES = temp3[n].text
        
        temp3 = temp2[5].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            TOTAL_EQUITY = temp3[n].text
        
        temp3 = temp2[17].find_all("td")
        for n in range(1,(len(temp4[0:-5]))):
            TRADE_PAYABLES = temp3[n].text
        
        
        
        Company_Names.append(company2)
        Total_Current_Liabilities.append(TOTAL_CURRENT_LIABILITIES)
        Year.append(year2)
        Total_Current_Assets.append(TOTAL_CURRENT_ASSETS)
        Inventories.append(INVENTORIES)
        Cash_and_Cash_Equivalents.append(CASH_AND_CASH_EQUIVALENTS)
        Current_Investments.append(CURRENT_INVESTMENTS)
        Short_Term_Borrowings.append(SHORT_TERM_BORROWINGS)
        Total_Assets.append(TOTAL_ASSETS)
        Total_Liabilities.append(TOTAL_LIABILITIES)
        Total_Non_Current_Assets.append(TOTAL_NON_CURRENT_ASSETS)
        Total_Non_Current_Liabilities.append(TOTAL_NON_CURRENT_LIABILITIES)
        Intangible_Assets.append(INTANGABLE_ASSETS)
        Fixed_Assets.append(FIXED_ASSETS)
        Capital_Employed.append(CAPITAL_EMPLOYED)
        Working_Capital.append(WORKING_CAPITAL)
        Trade_Receivables.append(TRADE_RECEIVABLES)
        Total_Equity.append(TOTAL_EQUITY)
        Trade_Payables.append(TRADE_PAYABLES)
        
        
        
    except Exception as e:
        print(e) 
            
    print(Company_Names)
    print(Year)
    print(Total_Current_Assets)
    print(Total_Current_Liabilities)
    print(Inventories)
    print(Cash_and_Cash_Equivalents)
    print(Current_Investments)
    print(Short_Term_Borrowings)
    print(Total_Assets)
    print(Total_Liabilities)
    print(Total_Non_Current_Assets)
    print(Total_Non_Current_Liabilities)
    print(Intangible_Assets)
    print(Fixed_Assets)
    print(Capital_Employed)
    print(Working_Capital)
    print(Trade_Receivables)
    print(Total_Equity)
    print(Trade_Payables)
    
    
    
```
