```python
# Final Code for finding the variables of company from balance sheet
    
moneycontrol_url = "https://www.moneycontrol.com/india/stockpricequote/"
letter = input("Please enter letter")
upper = letter[0].upper()
data_url = moneycontrol_url + upper
website_response = urReq(data_url)
letter_company_data = website_response.read()
beautified_results = bs(letter_company_data,"html.parser")
finding_box = beautified_results.find_all("div",{"class":"PT15"})
tr_finding_box = finding_box[0].find_all("tr")
    
for d in range(1,len(tr_finding_box)):
    td_finder = tr_finding_box[d].find_all("td")
    
if __name__ == "__main__":
    a_urls = company_urls()
    
    
company_url = a_urls[0:len(a_urls)]
Company_Names = []
Total_Current_Liabilities = []
Year = []


for l in company_url:
    try:
        company_page2 = requests.get(l)
        company_page2.encoding = "utf-8"
        beautified_company_page2 = bs(company_page2.text,"html.parser")
        detailed_financial2 = beautified_company_page2.find_all("div",{"class":"quick_links clearfix"})
        detailed_financial3 = detailed_financial2[0].find_all("li")

        balance_sheet2 = detailed_financial3[0].a["href"]
        balance_sheet_url2 = requests.get(balance_sheet2)
        balance_sheet_url2.encoding = "utf-8"
        beaurified_balance_sheet_url2 = bs(balance_sheet_url2.text,"html.parser")

        company_name2 = beaurified_balance_sheet_url2.find_all("div",{"class":"moneyprice_bx"})
        company2 = company_name2[0].h1.text


        balance_sheet_text2 = beaurified_balance_sheet_url2.find_all("div", {"class":"table-responsive financial-table"})
        temp2 = balance_sheet_text2[0].find_all("tr")
        temp3 = temp2[20].find_all("td")
        temp4 = temp2[0].find_all("td")
        for n in range(1,(len(temp4[0:-1]))):
            TOTAL_CURRENT_LIABILITIES=temp3[n].text
            year2 = temp4[n].text

        Company_Names.append(company2)
        Total_Current_Liabilities.append(TOTAL_CURRENT_LIABILITIES)
        Year.append(year2)
    
    except Exception as e:
        print(e) 
            
    print(Company_Names)
    print(Total_Current_Liabilities)
    print(Year)
            
```
